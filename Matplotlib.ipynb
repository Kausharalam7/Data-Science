**What is Matplotlib?**

Matplotlib is a Python library for creating static, interactive, and animated visualizations.

It is widely used in data analysis, machine learning, and scientific computing.

The most commonly used module is pyplot, which provides MATLAB-like plotting framework.

```python
import matplotlib.pyplot as plt
```

**What is a Line Chart?**

A line chart displays data points connected by straight lines.

Best for showing trends over time or continuous data.

Commonly used for time series analysis, sales trends, or performance monitoring.

```python
# line chart
x=[1,2,3,4,5]
y=[5,35,10,20,15]

plt.plot(x,y)
plt.show()
```

```python
x=['Mon','Tues','Wed','Thurs','Fri']
y=[23,19,10,20,5]

plt.plot(x,y)
plt.title("Bakery Sales this week")
plt.xlabel("Day of the week")
plt.ylabel("Sales per Day")
plt.show()
```

```python
# plt.plot(x,y,color='green',linestyle='line_Style',linewidth=value,marker='marker symbol',label='label name' )
#
months=[2,3,5,7]
sales=[1000,2000,1500,2500]

plt.plot(months,sales,color='green',linestyle='--',linewidth=5,marker='o', label='2025 sales data')
plt.xlabel("Month")
plt.ylabel("Sales Per Month")
plt.title("Monthly Sales Data Report")
plt.legend()
plt.grid(color='red',linestyle=":",linewidth=2)
plt.xlim(3,6)
plt.ylim(00,5000)
plt.xticks([2,3,5,7],['M1','M2','M3','M4'])
plt.show()
```

**What is a Bar Chart?**

A bar chart is used to represent categorical data with rectangular bars.

The length/height of the bar represents the value.

Commonly used for sales data, survey results, counts, or comparisons between categories.

```python
# plt.bar(x,height,color='color name',width=value,label='label name')
product=['A','B','C','D']
sales=[1000,1500,800,1200]

plt.bar(product,sales,color='orange',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```

**What is a Pie Chart?**

A pie chart represents data as slices of a circle.

Each slice shows the proportion (%) of each category relative to the total.

Useful for sales distribution, market share, or survey results.

```python
# plt.pie(values, labels=label_list,colors=colors_list,autopct='%1.1f%%')

regions=['North','South','East','West']
revenue=[3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%',colors=['gold','skyblue','red','coral'])
plt.title('Revenue Contributed By Region')
plt.show()
```

**What is a Histogram?**

A histogram is a graphical representation of the distribution of numerical data.

Data is divided into bins (intervals).

Height of each bin represents the count/frequency of values in that interval.

Useful for analyzing data distribution, outliers, skewness, and spread.

```python
#plt.hist(data, bins=number_of_bins,color='colorname',edgecolor='black')

scores= [45,76,23,78,56,87,23,45,21,47,96,85,76,82,19,29,82,37,65,95,14]
plt.hist(scores,bins=5,color='yellow',edgecolor='green')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')
plt.show()
```

**What is a Scatter Plot?**

A scatter plot displays individual data points on a 2D plane using (x, y) coordinates.

It is used to observe relationships, trends, or correlations between two variables.

Common in EDA (Exploratory Data Analysis) to spot patterns, clusters, or outliers.

```python
# Scatter PLot
# plt.scatter(x,y,color='color name',marker='marker style',label='label name')
hour_studies=[1,2,3,4,5,6,7,8]
exam_scores=[50,55,60,65,70,75,80,85]

plt.scatter(hour_studies,exam_scores,color='green',marker='o', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between Study Time and Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3],[50,55,60], color='blue',label='Class A') # g1
plt.scatter([1,2,3],[45,50,52], color='orange',label='Class B')  #g2
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparison of Two Class')
plt.legend()
plt.grid(True)
plt.show()
```

**What is a Subplot?**

A subplot is a plot inside a larger figure.

Allows you to compare multiple plots side by side or in a grid.

Useful for comparing different variables or datasets visually.

```python
# subplot
x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) # 1 row,1 col, 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) # 1 row,2nd col, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, color='blue')
plt.title('Line Plot 1')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, color='red')
plt.title('Line Plot 2')

plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.bar(['A','B','C','D'], [4,7,5,6], color='orange')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.scatter([1,2,3,4], [5,7,6,8], color='green')
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.hist([1,2,2,3,3,3,4,4,5], bins=5, color='purple')
plt.title('Plot 4')

plt.tight_layout()  # Automatically adjusts spacing
plt.show()
```

```python
# fig, ax=plt.subplots(nrow,ncols,figsize=(width,height))
fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y,color='black')
ax[0].set_title('Line Plot')
ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Chart')
fig.suptitle('Comparisons of Line and Bar Charts')
plt.tight_layout
plt.savefig('comparions bw two plot',dpi=300,bbox_inches='tight')
plt.show()
```
