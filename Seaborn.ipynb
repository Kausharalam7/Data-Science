Seaborn is a Python data visualization library built on top of Matplotlib. It makes creating beautiful, statistical graphics much simpler and faster.

Why Seaborn?

Built-in support for statistical plots (histograms, boxplots, regression plots).

Works directly with Pandas DataFrames.

Better aesthetics and defaults than Matplotlib.

One-liners for complex plots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
y_1=sns.load_dataset("penguins")
```

**Line Plot**

A line plot is mainly used to visualize trends over a continuous variable (often time, but can be anything numeric).

```python
sns.lineplot(
    x=None,           # Column for x-axis
    y=None,           # Column for y-axis
    hue=None,         # Different colors for categories
    style=None,       # Different line styles for categories
    markers=False,    # Add markers on data points
    data=None,        # DataFrame or dataset
    errorbar=('ci', 95),       # Confidence interval (None to remove, "sd" for standard deviation)
    palette=None,     # Custom color palette
    linewidth=2,      # Thickness of line
    dashes=True,      # Line dash pattern
    legend="auto",    # Show legend
)
```

```python
# matplotlib
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]
plt.plot(var,var1)
plt.show()
```

**Bar Plot**

A bar plot is used to visualize the relationship between a categorical variable (x) and a numeric variable (y).

Each bar’s height represents the mean (by default) of the numeric variable.

You can add error bars (confidence interval / standard deviation).

```python
sns.barplot(
    x=None,            # Categorical variable
    y=None,            # Numeric variable
    hue=None,          # Separate bars by category
    data=None,         # Dataset (DataFrame)
    errorbar=("ci",95),# Error bar: ("ci",95), "sd", or None
    palette=None,      # Colors
    order=None,        # Custom order for categories
    orient=None,       # "v" for vertical, "h" for horizontal
    estimator=np.mean, # Function to aggregate (default: mean)
)
```

```python
sns.barplot(x=y_1.island,y=y_1.bill_length_mm)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=y_1,hue="sex",order=order_1,hue_order=["Female","Male"],errcolor='y')
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=y_1,orient="h",color='r')
plt.show()
```

**Histogram (sns.histplot)**

A histogram shows the distribution of a single numeric variable.

It divides data into intervals (bins) and counts how many values fall in each bin.

You can also overlay Kernel Density Estimation (KDE) to visualize probability distribution.

```python
sns.displot(
    data=None,       # DataFrame or array
    x=None,          # Column name or numeric array
    y=None,          # (less common, for 2D histograms)
    hue=None,        # Separate histograms by category
    bins=None,       # Number of bins (default auto)
    binwidth=None,   # Width of each bin
    kde=False,       # Add a density curve
    multiple="layer",# Options: "layer", "stack", "dodge", "fill"
    element="bars",  # "bars" (default) or "step"
    stat="count",    # Options: "count", "frequency", "density", "probability"
)
```

```python
sns.displot(y_1['flipper_length_mm'])
plt.show()
```

```python
sns.displot(y_1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,color='r')
plt.show()
```

**Scatter Plot (sns.scatterplot)**

A scatter plot shows how two numerical variables relate to each other.

Each point represents one observation.

You can also add categorical separation (hue, style, size) to show extra dimensions.

```python
sns.scatterplot(
    data=None,      # DataFrame or array
    x=None,         # Column for X-axis
    y=None,         # Column for Y-axis
    hue=None,       # Different colors for categories
    style=None,     # Different marker shapes
    size=None,      # Different sizes of markers
    palette=None,   # Color palette
    alpha=None,     # Transparency (0 to 1)
)
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm', data=y_1,hue='sex',palette='gist_ncar',style='sex',markers=m)
plt.show()
```

**Heatmap (sns.heatmap)**

A heatmap is a graphical representation of data where values are represented by colors.

Very useful for:
* Correlation matrices (to check relationships between features)
* Pivot tables (to visualize aggregated values)
* Highlighting patterns in numerical data

```python
sns.heatmap(
    data,             # 2D dataset (NumPy array, DataFrame, correlation matrix)
    vmin=None, vmax=None,   # Value range for colors
    cmap="coolwarm",        # Colormap (e.g., "viridis", "plasma", "Blues", etc.)
    annot=False,            # Annotate each cell with numeric value
    fmt=".2f",              # Format of annotation (e.g., 2 decimals)
    linewidths=0.5,         # Space between cells
    linecolor="black",      # Color of lines between cells
    cbar=True,              # Show color bar
)
```

**Count Plot** (sns.countplot)

A count plot shows the count (frequency) of each category in a categorical column.

It’s basically a bar plot, but the height of the bars is automatically the count of observations.

Commonly used in EDA (Exploratory Data Analysis) to understand categorical distributions.

Use the hue parameter to split one category by another (e.g., gender split by smoker status).

```python
sns.countplot(
    data=None,        # DataFrame
    x=None, y=None,   # Column for categories (x for vertical, y for horizontal)
    hue=None,         # Sub-categories (color split)
    palette=None,     # Color scheme
    order=None,       # Order of categories
    hue_order=None,   # Order of hue categories
)
```

difference between countplot and barplot in Seaborn

countplot → counts occurrences automatically.

barplot → requires you to provide a numerical column to aggregate.


```python
 var=sns.load_dataset("tips")
 var.head(5)
```

```python
sns.countplot(x='sex',data=var)
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color='r')
plt.show()
```

**Violin Plot** (sns.violinplot)

A violin plot combines the features of a boxplot and a KDE (Kernel Density Estimation) plot.

It shows the distribution of data (like a histogram but smoothed).

Displays median, quartiles (like boxplot).

Useful for comparing distributions across categories.

```python
sns.violinplot(
    data=None,       # DataFrame
    x=None, y=None,  # Variables to plot
    hue=None,        # Sub-categories
    split=False,     # Split violins for hue categories
    inner="box",     # Show 'box', 'quartile', 'point', or 'stick'
    scale="width",   # Scale violins by 'area', 'count', or 'width'
    palette=None     # Color scheme
)
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,palette='Dark2')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=["Dinner","Lunch"],color='r',linewidth=3)
plt.show()
```

```python
sns.violinplot(y=var['total_bill'])
plt.show()
```

**Pair Plot** (sns.pairplot)

A pair plot creates a grid of scatterplots for every numerical feature against each other, with histograms (or KDE) on the diagonals.

It’s basically EDA in one line of code.

```python
sns.pairplot(
    data,             # DataFrame
    hue=None,         # Categorical variable for coloring
    vars=None,        # Subset of columns to plot
    kind="scatter",   # 'scatter', 'kde', or 'reg'
    diag_kind="auto", # 'auto', 'hist', 'kde'
    palette=None,     # Color scheme
    corner=False      # Show only lower triangle if True
)
```

```python
sns.pairplot(var,hue='sex')
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'])
plt.show()
```

**Strip Plot** (sns.stripplot)

* A strip plot displays individual data points for categorical variables as dots aligned along the category axis.
* Think of it as a scatter plot for categorical vs numerical data.

```python
sns.stripplot(
    data=None,
    x=None,
    y=None,
    hue=None,
    jitter=True,        # Adds small random noise to avoid overlap
    dodge=False,        # Separates points by hue levels
    palette=None,
    size=5,             # Marker size
    alpha=0.7           # Transparency
)
```

```python
var=sns.load_dataset("tips");
var
```

**Box Plot** (sns.boxplot)

A box plot (also called a whisker plot) shows the distribution of numerical data using:

Median (middle line inside box)

Quartiles (Q1, Q3) (the edges of the box)

Whiskers (range of non-outlier values)

Outliers (points outside whiskers)

```python
sns.boxplot(
    data=None,
    x=None,
    y=None,
    hue=None,
    orient=None,     # "v" (vertical) or "h" (horizontal)
    palette=None,
    showcaps=True,   # Show whisker caps
    fliersize=5,     # Size of outlier points
    linewidth=1.5
)
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,hue='sex')
plt.show()
```

**Factor Plot** (now CatPlot)

A factor plot is a general categorical plot that allows you to visualize data across categories using different types of plots (point, bar, box, violin, etc.) by just changing the kind parameter.

```python
sns.catplot(x="day", y="total_bill", data=var, kind="point")
plt.show()
```

```python
# In older versions of Seaborn (<0.9), there was a function called sns.factorplot().
# In newer versions, factorplot() has been renamed to sns.catplot().
```

**CatPlot**

sns.catplot() is a unified function in Seaborn for categorical plots.

It combines bar plot, box plot, violin plot, strip plot, swarm plot, point plot etc. into a single function.

Very useful for EDA (Exploratory Data Analysis) when analyzing relationships between categorical variables and numeric variables.

```python
sns.catplot(
    x=None,
    y=None,
    hue=None,
    col=None,
    row=None,
    data=None,
    kind="strip",   # Options: "strip", "swarm", "box", "violin", "bar", "count", "point"
    height=5,
    aspect=1
)
```

```python
sns.catplot(x='tip',y='size',data=var)
plt.show()
```

```python
sns.catplot(x='day',y='size',data=var,hue='sex',kind='bar',palette='Oranges')
plt.show()
```

```python
sns.catplot(x='day',y='size',data=var,hue='sex',kind='box',palette='Oranges')
plt.show()
```
